```python
import pandas as pd
import matplotlib.pyplot as plt
veri = pd.read_csv("C:\\Users\\cdggt\\Desktop\\50_Startups.csv")
```

```python
plt.scatter(veri["R&D Spend"], veri["Profit"], color="blue")
plt.xlabel("R&D Harcaması")
plt.ylabel("Kar")
plt.title("R&D Harcaması-Kar")
plt.show()
```

R&D maliyetini yatay , akrı dikey stuna yerleştiriyoruz ve arge çalışmalarıyla kar miktarının artışının paralel olduğunu gözlemliyoruz

```python
plt.scatter(veri["Administration"],veri["Profit"], color="green")
plt.xlabel("Yönetim Harcaması")
plt.ylabel("Kar")
plt.title("Yönetim Harcaması vs Kar")
plt.show()
```

Bu kısımda yönetim maliyetiyle kar arsında net bir ilişki gözlenmiyor.yönetim harcamalarının kar artısında tek basına belirgin bir etkisi yok.

```python
ort_kar = veri.groupby("State")["Profit"].mean()
ort_kar.plot(kind="bar", color=["blue", "yellow", "blue"])
plt.ylabel("Ortalama Kar")
plt.title("Eyaletlere Göre Ortalama Kar")
plt.show()
```

Floride eyaletindeki startup projeleri California ve New York eyaletlerine göre nispeten daha karlı oluyor.

```python
veri[["R&D Spend", "Administration", "Marketing Spend"]].plot(kind="box")
plt.title("Harcama Dağılımı")
plt.ylabel("Miktarı")
plt.show()
```

Boxplot grafiğide, R&D harcaması kısmında alt ve üst sınır arasındaki fark az, medyan değer 75.000 civarında
Yönetim harcamaları kısmında birbirine daha yakın değerler görünüyor,medyan değer 125.000 civarında
Pazarlama maliyetlerinde ise uç değerler birbirinden çok farklı, medyan değer 210.00
